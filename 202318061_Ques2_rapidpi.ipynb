```python
data = response.json()
data
```

```python
import pandas as pd
```

```python
title_ids = []
title_names = []
title_types = []
release_years = []
release_dates = []


for result in data['results']:
    title_ids.append(result['id'])
    title_names.append(result['titleText']['text'])
    title_types.append(result['titleType']['text'])
    release_years.append(result['releaseYear']['year'])
    release_date = result['releaseDate']
    release_dates.append(f"{release_date['day']}/{release_date['month']}/{release_date['year']}")


df = pd.DataFrame({
    'Title ID': title_ids,
    'Title Name': title_names,
    'Title Type': title_types,
    'Release Year': release_years,
    'Release Date': release_dates
})


print(df)
```
