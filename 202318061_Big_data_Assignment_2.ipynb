```python
df['text'] = df['text'].astype(str).apply(lambda x: re.sub(r'\d+|\n+|\)|\(|\@|\>|\!|\.|\-|\t|^\w\s|\|\:,', ' ', x))
df['text'] = df['text'].str.lower()
df
```

<h1> Question 1</h1>

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction import text
model = TfidfVectorizer(stop_words='english',max_features=1000)
```

```python
transformed_output=model.fit_transform(df['text'])
feature_dict=model.vocabulary_
len(feature_dict)
```

```python
print(transformed_output)
```

```python
feature_names=model.get_feature_names_out()
for words in feature_names:
    index_position=model.vocabulary_.get(words)
    idf_score=model.idf_[index_position]
    print(f'{feature_dict}:',idf_score)
    #print(f'{index_position}: ', idf_score)
```

```python
print(f'{index_position}: ', feature_names) 
```

<h1>Question 2</h1>

```python
df_tfidf = pd.DataFrame(transformed_output.toarray(), columns=model.get_feature_names_out())
df_tfidf
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(df_tfidf, df_tfidf)

#now cosaine similarity bw 0th and 2th sentence or document is -
cosine_sim
```

<h1>Question 3.</h1>

```python
new_query = df['text'][0]
new_query_vector = model.transform([new_query])
cosine_sim = pd.DataFrame(cosine_similarity(df_tfidf, new_query_vector), columns = ['cosine_similarity_score'], index = df['text'].index)
cosine_sim
```

```python
cosine_sim = cosine_sim.sort_values(by = ['cosine_similarity_score'], ascending=False)
cosine_sim
```

```python
#creating a mini recc sys to predicit top 10 most similar documents when given a query
output = df.loc[cosine_sim.index[0:10],:]
print(output['text'])
```
